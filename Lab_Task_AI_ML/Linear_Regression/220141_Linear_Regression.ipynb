**Data Preprocessing & Cleannig**

Drop unnecessary columns

```python
# These columns are not useful for prediction
cols_to_drop = ['Location', 'Date', 'Product ID', 'Brand', 'Storage Condition',
                'Production Date', 'Expiration Date', 'Customer Location', 'Farm Size']

df_clean = df.drop(columns=cols_to_drop)
print("Remaining columns:", df_clean.columns.tolist())
```

 Handle Missing Values

```python
print("Missing values before:\n", df_clean.isnull().sum())

# Fill numeric columns with median
numeric_cols = df_clean.select_dtypes(include='number').columns
df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].median())

print("\nMissing values after:\n", df_clean.isnull().sum())
```

Encoded Categorical Columns

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df_clean.select_dtypes(include='object').columns.tolist()
print("Categorical columns to encode:", cat_cols)

le = LabelEncoder()
for col in cat_cols:
    df_clean[col] = le.fit_transform(df_clean[col].astype(str))

print("\nDone! Shape:", df_clean.shape)
df_clean.head()
```

Define Feature & Target

```python
X = df_clean.drop(columns=['Approx. Total Revenue(INR)'])
y = df_clean['Approx. Total Revenue(INR)']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

 Normalize / Scale the Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaling done! Sample:\n", X_scaled[:3])
```

Split into Train & Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)

print("Training size:", X_train.shape)
print("Testing size:", X_test.shape)
```

**Train The  Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

print("Model trained successfully! ✅")
```

```python
y_pred = model.predict(X_test)

print("Sample Predictions vs Actual:")
for i in range(10):
    print(f"Predicted: {y_pred[i]:,.2f}  |  Actual: {y_test.values[i]:,.2f}")
```

Calculate Error Metrics (MAE, MSE,R^2)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("=" * 40)
print(f"  MAE  (Mean Absolute Error)  : {mae:,.2f}")
print(f"  MSE  (Mean Squared Error)   : {mse:,.2f}")
print(f"  RMSE (Root MSE)             : {rmse:,.2f}")
print(f"  R²   (R-Squared Score)      : {r2:.4f}")
print("=" * 40)
```

Actual VS Predicted Plot

```python
residuals = y_test.values - y_pred

plt.figure(figsize=(8,5))
plt.plot(residuals[:100], color='tomato', linewidth=1)
plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
plt.title('Residuals / Loss Curve (First 100 Samples)')
plt.xlabel('Sample Index')
plt.ylabel('Residual (Actual - Predicted)')
plt.tight_layout()
plt.show()
```
